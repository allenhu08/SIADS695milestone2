```python
RANDOM_SEED = 42
```

```python
train_df = pd.read_csv('data/WikiLarge_Train.csv')

train_df['length']=train_df['original_text'].str.split().str.len()

train_df.sample(10)
```

```python
X=train_df['original_text']
y=train_df['label']
    
X_train, X_dev, y_train, y_dev = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED)
```

```python
print('training data labels:\n', y_train.value_counts(), '\n')
print('dev data labels:\n', y_dev.value_counts())
```

```python
%%time
min_df=50
ngram_range=(1,3)

vectorizer = TfidfVectorizer(stop_words=None, lowercase=False, #token_pattern = r'\b[a-z]{3,12}\b', 
                                     min_df=min_df, ngram_range=ngram_range).fit(train_df['original_text'])

X_train_trans=vectorizer.transform(X_train)
X_dev_trans=vectorizer.transform(X_dev)
```

```python
X_train_trans.shape
```

```python
%%time

clf = LogisticRegression(solver= 'lbfgs', max_iter=10000, 
                                 random_state=RANDOM_SEED).fit(X_train_trans, y_train)
print(type(clf))

y_pred=clf.predict(X_dev_trans)
f1score = f1_score(y_dev, y_pred, average='macro')
accuracy=clf.score(X_dev_trans, y_dev)

print('f1_score=',f1score, '; accuracy=', accuracy)
print('precision=', precision_score(y_dev, y_pred))
print('recall=',recall_score(y_dev, y_pred))
print('conf_matrix: \n', confusion_matrix(y_dev, y_pred), '\n')
plot_precision_recall_curve(clf, X_dev_trans, y_dev)
plot_roc_curve(clf, X_dev_trans, y_dev)
```

```python
%%time

clf = RandomForestClassifier(random_state=RANDOM_SEED, max_depth=15,n_estimators=100).fit(X_train_trans, y_train)

print(type(clf))

y_pred=clf.predict(X_dev_trans)
f1score = f1_score(y_dev, y_pred, average='macro')
accuracy=clf.score(X_dev_trans, y_dev)

print('f1_score=',f1score, '; accuracy=', accuracy)
print('precision=', precision_score(y_dev, y_pred))
print('recall=',recall_score(y_dev, y_pred))
print('conf_matrix: \n', confusion_matrix(y_dev, y_pred), '\n')
plot_precision_recall_curve(clf, X_dev_trans, y_dev)
plot_roc_curve(clf, X_dev_trans, y_dev)
```

```python
%%time

clf = MultinomialNB().fit(X_train_trans, y_train)

print(type(clf))

y_pred=clf.predict(X_dev_trans)
f1score = f1_score(y_dev, y_pred, average='macro')
accuracy=clf.score(X_dev_trans, y_dev)

print('f1_score=',f1score, '; accuracy=', accuracy)
print('precision=', precision_score(y_dev, y_pred))
print('recall=',recall_score(y_dev, y_pred))
print('conf_matrix: \n', confusion_matrix(y_dev, y_pred), '\n')
plot_precision_recall_curve(clf, X_dev_trans, y_dev)
plot_roc_curve(clf, X_dev_trans, y_dev)
```

```python
%%time

clf = MLPClassifier(hidden_layer_sizes = [10], solver='lbfgs', #max_iter=10000,
                         random_state = RANDOM_SEED).fit(X_train_trans, y_train)
print(type(clf))

y_pred=clf.predict(X_dev_trans)
f1score = f1_score(y_dev, y_pred, average='macro')
accuracy=clf.score(X_dev_trans, y_dev)

print('f1_score=',f1score, '; accuracy=', accuracy)
print('precision=', precision_score(y_dev, y_pred))
print('recall=',recall_score(y_dev, y_pred))
print('conf_matrix: \n', confusion_matrix(y_dev, y_pred), '\n')
plot_precision_recall_curve(clf, X_dev_trans, y_dev)
plot_roc_curve(clf, X_dev_trans, y_dev)
```
