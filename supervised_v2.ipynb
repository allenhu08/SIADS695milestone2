```python
RANDOM_SEED=42
```

```python
def evaluate(clf, y_pred):
    

    f1score = f1_score(y_test, y_pred, average='macro')
    accuracy = clf.score(X_test, y_test)

    print('f1_score= ', f1score, '; accuracy= ', accuracy)
    print('precision= ', precision_score(y_test, y_pred))
    print('recall= ', recall_score(y_test, y_pred))
    print('conf_matrix: \n', confusion_matrix(y_test, y_pred), '\n')
    plot_precision_recall_curve(clf, X_test, y_test)
    plot_roc_curve(clf, X_test, y_test)
    
def create_pipeline(clf):
    return Pipeline(
        [
                  ('transform', preprocessor.get_transformer()),
                  ('classify', clf)
        ]
    )
```

```python
%%time

print(X_train.head())
clf = LogisticRegression(solver= 'lbfgs', max_iter=1000, 
                                 random_state=RANDOM_SEED)

pipeline = create_pipeline(clf)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

evaluate(pipeline, y_pred)
```

```python
%%time

clf = RandomForestClassifier(random_state=RANDOM_SEED, max_depth=15,n_estimators=100)

print(type(clf))


pipeline = create_pipeline(clf)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

evaluate(pipeline, y_pred)
```

```python
%%time

clf = MultinomialNB()
print(type(clf))

pipeline = create_pipeline(clf)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

evaluate(pipeline, y_pred)
```

```python
%%time

clf = MLPClassifier(hidden_layer_sizes = [10], solver='lbfgs', max_iter=1000,
                         random_state = RANDOM_SEED)

print(type(clf))

pipeline = create_pipeline(clf)
pipeline.fit(X_train, y_train)
y_pred = pipeline.predict(X_test)

evaluate(pipeline, y_pred)
```
